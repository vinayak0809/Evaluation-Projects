```python
x = df.drop('capacity_mw' , axis = 1)
y = df['capacity_mw']
```

```python
# Scaling x values 
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
x = sc.fit_transform(x)
```

```python
pd.DataFrame(x).head()
```

```python
pd.DataFrame(x).describe()
```

```python
# Data has been scaled between 0 to 1  
```

```python
# Splitting the data into training and testing
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# Importing all the necessary models from scikit learn library 
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
models = { 'RandomForestRegressor' : RandomForestRegressor(),
           'GradientBoostingRegressor' : GradientBoostingRegressor(),
           'LinearRegression' : LinearRegression(),
           'DecisionTreeRegressor' : DecisionTreeRegressor(),
           'KNeighborsRegressor' : KNeighborsRegressor()
}
```

```python
# Train each model 
```

```python
for model_name, model in models.items():
    model.fit(x_train, y_train)
```

```python
# Evaluate Model performance
```

```python
for model_name, model in models.items():
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    print(f'Model_name : {model_name}')
    print(f'{model_name} : MSE = {mse}, R2 = {r2}, MAE = {mae}')
```

```python
# It suggest that Random forest Regressor emerges as the best model with low Mean Squared Errorand Mean Absolute Error as well as High R2 Score 
```

```python
# Cross Validation
```

```python
from sklearn.model_selection import cross_val_score
for model_name, model in models.items():
    score = cross_val_score(model, x, y, cv = 5)
    print(f'Average_valiadation_score of {model_name} : {score.mean()}')
    print(f'Standarad_deviation_of_{model_name} : {score.std()}')
    print('=====================================================')
```

```python
# After Cross-Validation it seems that Gradient boosting has the highest Average validation score and lowest standarad deviation
```

```python
# Now its time for Hyperparameter tuning of Gradient Boosting Regressor
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
GB = GradientBoostingRegressor()
```

```python
parameters = {'n_estimators' : [50,100,200],
             'learning_rate' : [0.01,0.1,5],
             'max_depth' : [3,5,7],}
```

```python
grid_search = GridSearchCV(estimator = GB, param_grid = parameters, cv = 5, n_jobs = -1)
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
# FINAL MODEL
```

```python
best_params = grid_search.best_params_
final_model = GradientBoostingRegressor(**best_params)
final_model.fit(x_train, y_train)
y_pred = final_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'Model_name : {final_model}')
print(f'{final_model} : MSE = {mse}, R2 = {r2}, MAE = {mae}')
```
